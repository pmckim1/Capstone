```python
from docx.enum.text import WD_LINE_SPACING


#%%
class SummaryMaker(object):
    def make_difference_phrase(self, _from, _to):
        difference = _from - _to
        #print("From: ", _from, "To: ",  _to, "Heres the difference",_from - _to)
        # Exactly equal.
        if difference == 0:
            return "the same as"
        # Within 2.
        elif difference < 2 and difference > -2:
            return "nearly the same as"

        elif difference >= 2 and difference < 10:
            return "higher than"
        elif difference <= -2 and difference > -10:
            return "less than"

        elif difference >= 10:
            return "much higher than"
        elif difference <= -10:
            return "much less than"

    def make_comparative_phrase(self, other_country):
        phrase = "{other_country}'s voting coincidence rates with {country} for UNGA roll-call votes ranged between {country_low} percent and {country_high} percent."
        # df_other_country=self.comparative_coinc.query(f"OGCountry==@{other_country.upper()}")
        df_other_country=self.comparative_coinc[self.comparative_coinc["OGCountry"] == other_country.upper()]
        df_other_country=df_other_country[df_other_country["Country"] == self.country.upper()]

        df_other_country=df_other_country.drop(columns=['OGCountry'])
        country_maximum=df_other_country["Coincidence"].max()
        country_minimum=df_other_country["Coincidence"].min()

        return phrase.format(
            country = self.country,
            other_country = other_country,
            country_low=int(country_minimum*100),
            country_high=int(country_maximum*100),
        )

    def make_paragraph_for_two_maps(self, from_year, to_year):

        map_country_coincidences_2021 = self.map_country_coincidences[self.map_country_coincidences["Year"] == 2021]
        map_country_coincidences_2021 = map_country_coincidences_2021[map_country_coincidences_2021["CountryB"].str.lower() != self.country.lower()]
        #lowest_country = map_country_coincidences_2021[map_country_coincidences_2021["Coincidence"] == map_country_coincidences_2021["Coincidence"].min()]["CountryB"].iloc[0].title()
        #hightest_country = map_country_coincidences_2021[map_country_coincidences_2021["Coincidence"] == map_country_coincidences_2021["Coincidence"].max()]["CountryB"].iloc[0].title()
        ic_con_delta = self.make_delta_dataframe(from_year, to_year)
        largest_drop_country = ic_con_delta[ic_con_delta["Point Difference"] == ic_con_delta["Point Difference"].min()]["CountryB"].iloc[0].title()
        largest_increase_country = ic_con_delta[ic_con_delta["Point Difference"] == ic_con_delta["Point Difference"].max()]["CountryB"].iloc[0].title()


        usa_vs_other_country_difference = ic_con_delta[ic_con_delta["CountryB"] == "UNITED STATES"]["Point Difference"].iloc[0]

        def make_improvement_phrase(difference):
            # Exactly equal.
            if difference == 0:
                return "stayed the same"
            # Within 2.
            elif difference < 0.1 and difference > -0.1:
                return "is nearly the same"

            elif difference >= 0.1 and difference < 2:
                return "improved"
            elif difference <= -0.1 and difference > -2:
                return "decreased"

            elif difference >= 2:
                return "significantly improved"
            elif difference <= -2:
                return "significantly decreased"

        phrase = "These two maps show voting coincidence metrics from {country}'s perspective (in contrast to the previous visualizations in this packet which show the United States' perspective).  "\
                "The second map shows point change in voting coincidence from "\
                  "{from_year} through {to_year}."\
                  "  The United States and {country} rate {either_improved_or_decreased} compared to 2017, five years prior."
        return phrase.format(
            to_year=to_year,
            from_year=from_year,
            country=self.country,
            #lowest_country=lowest_country,
            #highest_country=hightest_country,
            largest_drop_country=largest_drop_country,
            largest_increase_country=largest_increase_country,
            either_improved_or_decreased=make_improvement_phrase(usa_vs_other_country_difference),

        )

    def make_voting_coin_text_us_centric(self):
      df = self.comparative_coinc.groupby(['OGCountry','Year']).mean().reset_index()
      df["Country"] = "Global"
      #print(df.head())
      df["Region"] = "Global"
      #print(df.head())
      # US Only
      df = df[df["OGCountry"] == "UNITED STATES"]
      df_new_global_average = df[[*self.comparative_coinc.columns]]

      df = self.comparative_coinc.groupby(['OGCountry','Year', "Region"]).mean().reset_index()
      df["Country"] = df["Region"]
      print(df.head())
      # US Only
      df = df[df["OGCountry"] == "UNITED STATES"]
      df_new_regional_average = df[[*self.comparative_coinc.columns]]

      # df = pd.concat([df_new_regional_average, self.comparative_coinc, df_new_global_average])
      # df = df.query('OGCountry=="UNITED STATES"')
      # regional_average = df.query('Country==@self.region')
      # global_average = df.query('Country=="Global"')

      # print(df_new_global_average)
      # print(df_new_regional_average)
      regional_low = int(df_new_regional_average["Coincidence"].min()*100)
      regional_high = int(df_new_regional_average["Coincidence"].max()*100)
      global_low = int(df_new_global_average["Coincidence"].min()*100)
      global_high = int(df_new_global_average["Coincidence"].max()*100)

      # This section is copied from make_comaprative_phrase, to change the values it uses.
      df_other_country=self.comparative_coinc[self.comparative_coinc["OGCountry"] == "United States".upper()]
      df_other_country=df_other_country[df_other_country["Country"] == self.country.upper()]
      df_other_country=df_other_country.drop(columns=['OGCountry'])
      country_maximum=df_other_country["Coincidence"].max()
      country_minimum=df_other_country["Coincidence"].min()
      country_low=int(country_minimum*100)
      country_high=int(country_maximum*100)

      return "This line graph shows the change in {country}’s UNGA coincidence rate with the United States compared to the "\
              "global and regional coincidence rates from 2002 through 2021.  For this time frame "\
              "the United States' voting coincidence rates with {country} for UNGA roll-call votes ranged between {country_low} percent and {country_high} percent."\
              .format(
                  country=self.country,
                  region=self.region,
                  regional_low = regional_low,
                  regional_high = regional_high,
                  global_low = global_low,
                  global_high = global_high,
                  country_low = min(country_low, self.saved_coinc_percent),
                  country_high = max(country_high, self.saved_coinc_percent),
              )

    def make_document(self):
        document = docx.Document()
        font = document.styles['Normal'].font
        font.name = 'Times New Roman'
        font.size = Pt(12)
        #add_header(document, '{} United Nations Metrics'.format(self.country), 0)
        paragraph = document.add_paragraph(gsm.make_phrase())
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE
        #paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER

        # bargraph with overall coincidences
        add_header(document, 'Global, Regional, and {} Comparative Voting Coincidence with the United States - 2021'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_locality_bargraph_otherway()), width=Inches(6.5))
        paragraph = document.add_paragraph("(SBU) Note:  Because the UN acts on diverse issues, the voting record of a UN member state during UNGA sessions provides insight into a country’s policy orientation.  While most bodies in the UN system approve resolutions by consensus without votes being taken, General Assembly resolutions often require a “roll-call” vote, in part due to their political nature and non-binding aspect.  “Important” resolutions are defined by Public Law 101-246 as “votes on issues which directly affected important United States interests and on which the United States lobbied extensively.”  “Israel-related” issues refer to any resolution specifically mentioning the state or territory of Israel.  The titles and context of these resolutions usually remain the same each year, with an annual average of 16-20 one-sided resolutions against Israel.")
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE

        #bargraph of absentee rates
        add_header(document, 'Global, Regional, and {} Absentee Rates - 2021'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_locality_bargraph_otherway_absentee()), width=Inches(6.5))
        paragraph = document.add_paragraph("This chart shows the number of roll call votes in which {} did not register a vote.".format(self.country))
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE

        # 2022 UNGA heatmap
        add_header(document, 'United Nations General Assembly Voting Coincidence {} and the United States on Contested Resolutions to Date 2022'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_unga_vote_heatmap()), width=Inches(6))
        p.alignment = WD_ALIGN_PARAGRAPH.CENTER
        paragraph = document.add_paragraph("This heatmap above shows the votes of {} compared to the United States on all contested resolutions through July 2022.  The majority of contested resolutions are traditionally voted on in the last quarter of the year. Dark blue indicates the country voted with the United States, Red indicates it voted opposite of the United States, White indicates it abstained or did not vote.".format(self.country))
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE



        add_header(document, 'United Nations General Assembly Voting Coincidence {} and the United States 2002-2021'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_voting_coinc_graph_us_centric()), width=Inches(6.5))
        p.alignment = WD_ALIGN_PARAGRAPH.CENTER
        paragraph = document.add_paragraph(self.make_voting_coin_text_us_centric())
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE

        add_header(document, 'United Nations General Assembly Voting Coincidence {} and China, Russia, and the United States 2002-2021'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_voting_coinc_graph(["United States", "China","Russia"])), width=Inches(6.5))
        p.alignment = WD_ALIGN_PARAGRAPH.CENTER

        paragraph = document.add_paragraph("(SBU) The line graph above shows that for the same time frame, " + self.make_comparative_phrase("Russia") +
                                           " While " + self.make_comparative_phrase("China"))
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE

        #add_header(document, 'Locality Comparative Voting Coincidence - 2021'.format(self.country), level=1)
        #p = document.add_picture(bufferize_figure(gsm.make_locality_bargraph_otherway()), width=Inches(7))
        p.alignment = WD_ALIGN_PARAGRAPH.CENTER
        #add_header(document, 'United Nations Secretariat', level=1)
        #self.make_un_table(document)
        add_header(document, 'United Nations General Assembly {} and the United States Thematic Voting Coincidence 2017-2021'.format(self.country), level=1)
        # Every figure is added the same way:
        p = document.add_picture(bufferize_figure(gsm.make_issue_code_graphs()), width=Inches(10.5))
        p.alignment = WD_ALIGN_PARAGRAPH.CENTER

        paragraph = document.add_paragraph(f"These six graphs show the voting coincidence between the United States and {self.country} on six different important issues frequently voted on at UNGA.")
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE

        add_header(document, 'United Nations General Assembly Voting Coincidence between {} and All Other Countries 2021'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_first_map()), width=Inches(5))
        p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER
        add_header(document, 'Change in United Nations General Assembly Voting Coincidence between {} and All Other Countries 2017-2021'.format(self.country), level=1)
        p = document.add_picture(bufferize_figure(gsm.make_second_map(2017, 2021)), width=Inches(5))
        #p.alignment = WD_ALIGN_PARAGRAPH.CENTER
        p.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

        paragraph = document.add_paragraph(self.make_paragraph_for_two_maps(2017, 2021))
        paragraph.paragraph_format.line_spacing_rule = WD_LINE_SPACING.SINGLE
        return document

    def make_voting_coinc_graph(self, all_the_peer_countries):
        other_country_dfs = {}
        for other_country in all_the_peer_countries:
            df_other_country=self.comparative_coinc[self.comparative_coinc["OGCountry"] == other_country.upper()]
            df_other_country=df_other_country[df_other_country["Country"] == self.country.upper()]
            other_country_dfs[other_country] = df_other_country

        data = list()
        for country, country_df in other_country_dfs.items():
            data.append(
                go.Scatter(name=country, x=country_df["Year"], y=country_df["Coincidence"], mode='lines'),
            )
        #template="plotly"
        layout = go.Layout(
          margin=go.layout.Margin(
                #l=0, #left margin
                #r=0, #right margin
                #b=0, #bottom margin
                t=0, #top margin
            )
        )
        fig = go.Figure(data=data, layout=layout)
        fig.layout.height = 400
        fig.layout.width = 1000
        # Change the bar mode
        fig.update_layout(yaxis_tickformat='.0%', yaxis_range=[0,1])
        fig.update_layout(
         yaxis = dict(
            tickfont=dict(size=20)
          )
        )
        fig.update_layout(
          xaxis = dict(
            tickfont = dict(size=20)
          )
        )
        fig.update_layout(xaxis_title="Year", yaxis_title="Voting Coincidence")
        fig.update_traces(line=dict(width=5))
        fig.update_layout(
          autosize=True,
          margin = {'l':0,'r':0,'t':0,'b':0},
        )
        fig.update_layout({
          'plot_bgcolor': 'rgba(218, 223, 225, 1)',
          'paper_bgcolor': 'rgba(0, 0, 0, 0)',
        })
        fig.update_layout(
            legend=dict(
                title_font_family="Times New Roman",
                font=dict(size= 20)
            )
        )
        # fig.write_image(outfilename, width=800, height=600, engine="kaleido")
        return fig

    def make_voting_coinc_graph_us_centric(self):
        df = self.comparative_coinc.groupby(['OGCountry','Year']).mean().reset_index()
        df["Country"] = "Global"
        df["Region"] = "Global"
        df_new_global_average = df[[*self.comparative_coinc.columns]]

        df = self.comparative_coinc.groupby(['OGCountry','Year', "Region"]).mean().reset_index()
        df["Country"] = df["Region"]
        df_new_regional_average = df[[*self.comparative_coinc.columns]]


        df = pd.concat([df_new_global_average, df_new_regional_average, self.comparative_coinc])

        country_Region=[self.country.upper(), self.region, "Global"]
        df = df.query('Country==@country_Region')
        df = df.query('OGCountry=="UNITED STATES"')
        df['Country'] = df['Country'].apply(lambda x: x.title())
        fig = px.line(df, x="Year", y="Coincidence", color="Country", title="United Nations General Assembly Voting Coincidence China and the United States 2002-2021")

        #layout = go.Layout(
         #     margin=go.layout.Margin(
                    #t=0, #top margin
               # )
            #)
        fig.layout.height = 400
        fig.layout.width = 800
                # Change the bar mode
        fig.update_layout(yaxis_tickformat='.0%')#, font=dict(size=16,))
        fig.update_layout(
          yaxis = dict(
            tickfont = dict(size=18)
          )
        )
        fig.update_layout(
          xaxis = dict(
            tickfont = dict(size=18)
          )
        )
        fig.update_layout(yaxis_range=[0,1], xaxis_title="Year", yaxis_title="Voting Coincidence")
        fig.update_layout(
        yaxis = dict(
        dtick = 0.1
        ))
        fig.update_traces(line=dict(width=5))
        fig.update_layout(
          #autosize=True,
          #margin = {'l':0,'r':0,'t':0,'b':0},
        )
        fig.update_layout(
        title="UNGA Voting Coincidence Israel and US 2002-2021",
        font=dict(
        size=20,
    ))
        fig.update_layout(legend={'title_text':''})

        fig.update_layout({
                  'plot_bgcolor': 'rgba(218, 223, 225, 1)',
                  'paper_bgcolor': 'rgba(0, 0, 0, 0)',
                })
        fig.update_layout(
            legend=dict(
                title_font_family="Times New Roman",
                font=dict(size=18)
            )
        )
        return fig



    def make_locality_bargraph(self):
        my_new_df = pd.concat([
          self.global_coincidences.iloc[0].to_frame().transpose().rename(columns={"Region": "Locality"}),
          self.region_coincidences.iloc[0].to_frame().transpose().rename(columns={"Region": "Locality"}),
          self.country_coincidences.to_frame().transpose().rename(columns={"Country": "Locality"}),
        ])
        my_new_df = my_new_df.drop(columns=["Region"])
        my_new_df = my_new_df.drop(columns=[col_name for col_name in my_new_df.columns if col_name.startswith("Absentee")])
        #print(my_new_df)

        data = []
        for vote_type in my_new_df.columns:
          if vote_type == "Locality":
            continue
          data.append(
            go.Bar(name=vote_type, x=my_new_df["Locality"], y=my_new_df[vote_type], text=my_new_df[vote_type], textposition='auto')
          )

        fig = go.Figure(data=data)
        # Change the bar mode
        fig.update_layout(barmode='group')
        fig.layout.height = 250
        fig.layout.width = 500
        fig.update_layout(yaxis_tickformat='.0%')
        fig.update_layout(title_text='Global and Regional Group Average')

        fig.update_layout(
          autosize=True,
          margin = {'l':0,'r':0,'t':0,'b':0},
          #height = 300
        )
        return fig

    def make_locality_bargraph_otherway(self):
        my_new_df = pd.concat([
          self.global_coincidences.iloc[0].to_frame().transpose().rename(columns={"Region": "Locality"}),
          self.region_coincidences.iloc[0].to_frame().transpose().rename(columns={"Region": "Locality"}),
          self.country_coincidences.to_frame().transpose().rename(columns={"Country": "Locality"}),
        ])
        my_new_df = my_new_df.drop(columns=["Region"])
        my_new_df = my_new_df.drop(columns=[col_name for col_name in my_new_df.columns if col_name.startswith("Absentee")])
        # Transpose it.
        my_new_df = my_new_df.transpose()
        # Get ride of useless indexing.
        my_new_df = my_new_df.reset_index()
        # Make the top row the col titles.
        my_new_df.columns = my_new_df.iloc[0]
        # Eliminate the now-titled top row.
        my_new_df = my_new_df.iloc[1:]
        # The "Locality" was formerly the column title, make it the appropriate label of "Vote Type"
        my_new_df = my_new_df.rename(columns={"Locality": "Vote Type"})
        my_new_df['Vote Type'] = my_new_df['Vote Type'].str.replace('Voting Coincidence','')

        data = []
        for locality in my_new_df.columns:
          if locality == "Vote Type":
            continue
          data.append(
            go.Bar(name=locality, x=my_new_df["Vote Type"], y=my_new_df[locality], text=[f'{i}' for i in my_new_df[locality]], textposition='auto')
          )

        fig = go.Figure(data=data)
        # Change the bar mode
        fig.update_layout(barmode='group')
        fig.layout.height = 350
        fig.layout.width = 800

        fig.update_layout(yaxis_tickformat='.0%')
        #fig.update_layout(
        #xaxis = dict(
       # tickfont = dict(size=20)))
        fig.update_layout(xaxis_title="Vote Type", yaxis_title="Voting Coincidence", font=dict(size=14))
        fig.update_layout({
        'plot_bgcolor': 'rgba(218, 223, 225, 1)',
        'paper_bgcolor': 'rgba(0, 0, 0, 0)',
         })

        #fig.update_layout(title_text='Global and Regional Group Average')



        fig.update_layout(
          autosize=True,
         margin = {'l':0,'r':0,'t':0,'b':0},
        )
        return fig


    def make_locality_bargraph_otherway_absentee(self):
        my_new_df = pd.concat([
          self.global_coincidences.iloc[0].to_frame().transpose().rename(columns={"Region": "Locality"}),
          self.region_coincidences.iloc[0].to_frame().transpose().rename(columns={"Region": "Locality"}),
          self.country_coincidences.to_frame().transpose().rename(columns={"Country": "Locality"}),
        ])
        my_new_df = my_new_df.drop(columns=["Region"])
        my_new_df = my_new_df.drop(columns=[col_name for col_name in my_new_df.columns if col_name.startswith("Voting Coincidence")])
        # Transpose it.
        my_new_df = my_new_df.transpose()
        # Get ride of useless indexing.
        my_new_df = my_new_df.reset_index()
        # Make the top row the col titles.
        my_new_df.columns = my_new_df.iloc[0]
        # Eliminate the now-titled top row.
        my_new_df = my_new_df.iloc[1:]
        #print(my_new_df)
        # The "Locality" was formerly the column title, make it the appropriate label of "Vote Type"
        my_new_df = my_new_df.rename(columns={"Locality": "Absentee Type"})
        my_new_df['Absentee Type'] = my_new_df['Absentee Type'].str.replace('Absentee Rate','')

        data = []
        for locality in my_new_df.columns:
          if locality == "Absentee Type":
            continue
          data.append(
            go.Bar(name=locality, x=my_new_df["Absentee Type"], y=my_new_df[locality], text=[f'{i}' for i in my_new_df[locality]], textposition='auto')
          )

        fig = go.Figure(data=data)
        # Change the bar mode
        fig.update_layout(barmode='group')
        fig.layout.height = 350
        fig.layout.width = 800

        fig.update_layout(yaxis_tickformat='.0%', yaxis_range=[0,1])
        #fig.update_layout(
        #xaxis = dict(
       # tickfont = dict(size=20)))
        fig.update_layout(xaxis_title="Absentee Type", yaxis_title="Absentee Rate", font=dict(size=14))
        fig.update_layout({
        'plot_bgcolor': 'rgba(218, 223, 225, 1)',
        'paper_bgcolor': 'rgba(0, 0, 0, 0)',
         })

        #fig.update_layout(title_text='Global and Regional Group Average')



        fig.update_layout(
          autosize=True,
         margin = {'l':0,'r':0,'t':0,'b':0},
        )
        return fig


    def make_short_document(self):
        document = docx.Document()
        self.make_un_table(document)
        return document

    def make_first_map(self):
        map_country_coincidences_2021 = self.map_country_coincidences[self.map_country_coincidences["Year"] == 2021]
        fig = px.choropleth(map_country_coincidences_2021, locations="CountryB",
                            color="Coincidence",
                            hover_name="CountryA",
                            locationmode="country names",
                            #title=f"{self.country} Voting Coincidence",
                            color_continuous_scale=px.colors.sequential.RdBu)
        fig.update_layout(yaxis_tickformat='%' )
        fig.update_layout(xaxis_tickformat='.0%' )
        fig.layout.height = 400
        fig.layout.width = 700
        fig.update_layout(
          autosize=True,
          margin = {'l':0,'r':0,'t':0,'b':0},
        )
        return fig

    def make_delta_dataframe(self, from_year, to_year):
        ic_con_from = self.map_country_coincidences[self.map_country_coincidences["Year"] == from_year].drop(columns=["Year"]).rename(columns={"Coincidence": "Voting_Coincidence_from"})
        ic_con_to = self.map_country_coincidences[self.map_country_coincidences["Year"] == to_year].drop(columns=["Year"]).rename(columns={"Coincidence": "Voting_Coincidence_to"})
        ic_con_delta = ic_con_to.merge(ic_con_from, how="left", left_on=["CountryA", "CountryB"], right_on=["CountryA", "CountryB"])
        ic_con_delta["Point Difference"] = ic_con_delta.apply(lambda x: x["Voting_Coincidence_to"] - x["Voting_Coincidence_from"], axis=1)
        # ic_con_delta = ic_con_delta.drop(columns=["Voting_Coincidence_from", "Voting_Coincidence_to"])
        ic_con_delta = ic_con_delta.round(2)
        return ic_con_delta

    def make_second_map(self, from_year, to_year):
        ic_con_delta = self.make_delta_dataframe(from_year, to_year)
        fig = px.choropleth(ic_con_delta, locations="CountryB",
                            color="Point Difference",
                            hover_name="CountryB",
                            locationmode="country names",
                            #title=f"{self.country} Voting Coincidence Change {from_year} - {to_year}",
                            color_continuous_scale=px.colors.sequential.RdBu)
        fig.update_layout(yaxis_tickformat='.0%' )
        fig.update_layout(xaxis_tickformat='.0%' )
        fig.layout.height = 400
        fig.layout.width = 700
        fig.update_layout(
          autosize=True,
          margin = {'l':0,'r':0,'t':0,'b':0},
        )
        fig.update_layout({
          'plot_bgcolor': 'rgba(218, 223, 225, 1)',
          'paper_bgcolor': 'rgba(0, 0, 0, 0)',
        })
        return fig

    def make_issue_code_graphs(self):
        df = self.issue_code_coin_us_cn_ru.groupby(['CountryA', 'Issue_Code', 'Year']).mean().reset_index()
        df["CountryB"] = "Global"
        df["UN_Regional_Groups"] = "Global"
        ic_con_new_global_average = df[[*self.issue_code_coin_us_cn_ru.columns]]

        df = self.issue_code_coin_us_cn_ru.groupby(['CountryA', 'Issue_Code', 'Year', "UN_Regional_Groups"]).mean().reset_index()
        df["CountryB"] = df["UN_Regional_Groups"]
        ic_con_new_regional_average = df[[*self.issue_code_coin_us_cn_ru.columns]]

        df = pd.concat([ic_con_new_global_average, ic_con_new_regional_average, self.issue_code_coin_us_cn_ru])
        df["Issue_Code"]=df['Issue_Code'].replace({'Palestinian Conflict': 'Israel Palestinian Issues'})
        #print(df.Issue_Code.unique())
        issue_code_entry = ["Human Rights", "Arms Treaties", "Sustainable Development", "Nuclear Weapons", "Israel Palestinian Issues", "Decolonialism"]

        df = df[df['CountryB'].isin(["Global", self.region, self.country.upper()])]
        df = df[df['Issue_Code'].isin(issue_code_entry)]
        df = df[df['CountryA'] == "UNITED STATES"]
        df['CountryB'] = df['CountryB'].apply(lambda x: x.title())

        fig = px.line(
            df,
            x="Year",
            y="Voting_Coincidence",
            color='CountryB',
            facet_col="Issue_Code",
            facet_col_wrap=2,
            facet_row_spacing=0.05,
            facet_col_spacing=0.07,
        )
        fig.layout.height = 1000
        fig.layout.width = 1200
        fig.update_xaxes(tick0=2017, dtick=1)
        fig.update_layout(yaxis_title=None)
        fig.update_layout(yaxis_range=[0,1])
        fig.update_layout(xaxis_title=None)
        fig.update_layout({
            'plot_bgcolor': 'rgba(218, 223, 225, 1)',
            'paper_bgcolor': 'rgba(0, 0, 0, 0)',
        })
        fig.update_layout(
        yaxis = dict(
        tickfont = dict(size=20)))
        fig.update_xaxes(tickfont_size=20)
        fig.update_yaxes(tickfont_size=20)
        fig.update_yaxes(tickformat='.0%')
        #fig.update_yaxes(row=1, yaxis_tickformat='.0%')
        #fig.update_yaxes1(yaxis_tickformat='.0%')
        fig.update_layout(legend={'title_text':''})
        for annotation in fig.layout.annotations:
          annotation.text = annotation.text.split("=")[1]
        fig.update_annotations(font=dict(size=34))
        fig.update_traces(line=dict(width=5))
        fig.for_each_annotation
        fig.update_layout(
          legend=dict(
            font=dict(
            size=18,
            color="black"
            )
          )
        )
        return fig

    def __init__(self, country, country_coincidences, region_coincidences, comparative_coinc, map_country_coincidences, issue_code_coin, votes_unga):
        self.country = country
        print(self.country)
        self.country_coincidences = country_coincidences[country_coincidences["Country"] == country].iloc[0]
        self.region = self.country_coincidences["Region"]
        self.region_coincidences = region_coincidences[region_coincidences["Region"] == self.region]
        self.global_coincidences = region_coincidences[region_coincidences["Region"] == "Global"]

        self.comparative_coinc = comparative_coinc
        self.map_country_coincidences = map_country_coincidences.reset_index(drop=True)
        self.map_country_coincidences = self.map_country_coincidences[self.map_country_coincidences["CountryA"].str.lower() == self.country.lower()]
        self.issue_code_coin_us_cn_ru = issue_code_coin

        # Some of these have extra spaces in their words.
        self.comparative_coinc["Region"] = self.comparative_coinc.apply(lambda row: ' '.join(row["Region"].split()).strip(), axis=1)
        # Remove everything after the first "(" in the region, f.x. "Western European & Other (Weog)"
        self.comparative_coinc["Region"] = self.comparative_coinc.apply(lambda row: row["Region"].split("(", 1)[0].strip(), axis=1)
        self.region_coincidences["Region"] = self.region_coincidences.apply(lambda row: row["Region"].split("(", 1)[0].strip(), axis=1)

        # Cherry pick this country's region from the first row of the Region Coincidences df.
        self.region = self.region_coincidences.iloc[0, 0]

        # Needed for the phrase maker.
        country_coincidences_without_global = country_coincidences[country_coincidences["Country"] != "GLOBAL AVERAGE"]
        country_coincidences_without_global = country_coincidences_without_global.sort_values('Country', ascending=True)
        country_coincidences_without_global = country_coincidences_without_global.sort_values('Voting Coincidence Roll-Call Votes', ascending=False, kind='stable')
        # print(country_coincidences_without_global[['Country', 'Voting Coincidence Roll-Call Votes']])
        self.number_of_countries = len(country_coincidences_without_global)
        self.us_coincidence_rank = np.where([country_coincidences_without_global["Country"] == self.country])[1][0] + 1
        self.votes_unga = votes_unga[votes_unga['Country'] == self.country]

        self.saved_coinc_percent = None


    def make_unga_vote_heatmap(self):
        df_country=self.votes_unga.round(2)
        df_country=df_country.set_index("Country")
        df_country = df_country.drop('Total Coincidence ', axis=1)

        fig = px.imshow(df_country,  text_auto=True, aspect="auto", color_continuous_scale="RdBu", zmin=0, zmax=1)#, title="UNGA Africa Voting Coincidence for Recent Ukraine Votes")
        fig.layout.height = 400
        fig.layout.width = 800
        fig.update_xaxes(side="top")
        fig.update_yaxes(tickfont_family="Arial Black")
        fig.update_xaxes(tickfont_family="Arial Black")
        fig.update_xaxes(tickfont_size=16)
        fig.update_yaxes(tickfont_size=16)
        fig.update_layout(yaxis_title=None)
        return fig


    def make_phrase(self):
        coinc_percent = self.country_coincidences["Voting Coincidence Roll-Call Votes"]*100
        self.saved_coinc_percent = int(coinc_percent)
        #print(self.region_coincidences["Voting Coincidence Roll-Call Votes"])
        regional_coinc_percent = float(self.region_coincidences["Voting Coincidence Roll-Call Votes"])*100
        global_coinc_percent= float(self.global_coincidences["Voting Coincidence Roll-Call Votes"])*100
        coinc_important_percent = float(self.country_coincidences["Voting Coincidence Important"])*100
        regional_coinc_important_percent = float(self.region_coincidences["Voting Coincidence Important"])*100
        global_coinc_important_percent = float(self.global_coincidences["Voting Coincidence Important"])*100

        phrase = "\n(SBU) Summary: {country:}'s voting coincidence with the United States at the UN General Assembly (UNGA) "\
            "on contested (i.e. “roll-call” votes) resolutions in 2021 was {coinc_percent:g} percent, {difference_phrase_region_overall:} the {regional_coinc_percent:g} percent {region:} "\
            "group average, and {difference_phrase_global_overall:} the global average of {global_coinc_percent:g} percent.  For votes "\
            "specified as “important” {country:}'s voting coincidence with the United States was "\
            "{coinc_important_percent:g} percent.  "\
            "This coincidence is {difference_phrase_region_important:} the regional average of {regional_coinc_important_percent:g} percent and {difference_phrase_global_important:} the "\
            "global average of {global_coinc_important_percent:g} percent." \
            "  {country:}'s voting coincidence with the United States was {ranking:} highest out of {number_of_countries:} countries."
        return phrase.format(
            country=self.country,
            coinc_percent=coinc_percent,
            regional_coinc_percent=regional_coinc_percent,
            region=self.region,
            difference_phrase_region_overall = self.make_difference_phrase(coinc_percent, regional_coinc_percent),
            difference_phrase_global_overall= self.make_difference_phrase(coinc_percent, global_coinc_percent),
            global_coinc_percent=global_coinc_percent,
            coinc_important_percent=coinc_important_percent,
            regional_coinc_important_percent=regional_coinc_important_percent,
            global_coinc_important_percent=global_coinc_important_percent,
            difference_phrase_region_important = self.make_difference_phrase(coinc_important_percent, regional_coinc_important_percent),
            difference_phrase_global_important = self.make_difference_phrase(coinc_important_percent, global_coinc_important_percent),

            russia_compare = self.make_comparative_phrase("Russia"),
            china_compare = self.make_comparative_phrase("China"),
            us_compare = self.make_comparative_phrase("United States"),
            ranking = self.us_coincidence_rank,
            number_of_countries = self.number_of_countries,
        )

def add_header(document, text, level):
    h=document.add_heading(level=level)
    h.add_run(text).font.name='Times New Roman'
    h.style.font.color.rgb = RGBColor(0,0,0)
    h.alignment = WD_ALIGN_PARAGRAPH.CENTER
    return h

def bufferize_figure(fig):
    img_buffer = io.BytesIO()
    fig.write_image(img_buffer, scale=0)#, width=800, height=600)
    img_buffer.seek(0)
    return img_buffer

os.makedirs("./output/", exist_ok=True)
for country in ["United Kingdom"]:
    gsm = SummaryMaker(country, df_country_coinc, df_regional_coinc, df_us_china_rus, df_coin_map, ic_con_new, votes_2022)
    #gsm.make_document().save('./output/Voting_Coincidence_{}_9_6_22.docx'.format(country))
    # print(gsm.make_paragraph_for_two_maps(2017, 2021))
    gsm.make_voting_coinc_graph_us_centric().show()
    # gsm.make_locality_bargraph_otherway_absentee().show()
    # print(gsm.make_voting_coin_text_us_centric())
    #gsm.make_first_map().show()
    #gsm.make_second_map(2002,2021).show()
    gsm.make_issue_code_graphs().show()
    #print(gsm.make_phrase())
    #print(gsm.make_paragraph_for_two_maps(2017,2021))
    #gsm.make_unga_vote_heatmap().show()

```
